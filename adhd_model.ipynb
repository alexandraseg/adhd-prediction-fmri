# Introduction, Objective and Dataset Description

## Problem

Neuropsychiatric disorders that occur in development, like anxiety, depression, autism, and attention deficit hyperactivity disorder, or ADHD, often differ in how and to what extent they affect males and females. ADHD occurs in about 11% of adolescents, with around 14% of boys and 8% of girls having a diagnosis. There is some evidence that girls with ADHD can often go undiagnosed, as they tend to have more inattentive symptoms which are harder to detect. Girls with ADHD who are undiagnosed will continue suffering with symptoms that burden their mental health and capacity to function.

For more domain context see the following videos: The Role of AI in Advancing Women's Brain Health Research [1] and Sex-specific differences in the healthy and disordered brain [2].

## Goal

Build a model to predict both an individual‚Äôs sex and their ADHD diagnosis using functional brain imaging data of children and adolescents and their socio-demographic, emotions, and parenting information.

Challenge Question and Task:
‚ÄúWhat brain activity patterns are associated with ADHD; are they different between males and females, and, if so, how?‚Äù

The task is to create a multi-outcome model to predict two separate target variables: 1) ADHD (1=yes or 0=no) and 2) female (1=yes or 0=no).

#### üë™ Alabama Parenting Questionnaire (APQ) ‚Äì Question Format

Each of the following items in the APQ assesses a specific aspect of parenting behavior. Responses are rated on a 5-point Likert scale indicating the frequency of the behavior:

**Answer Scale:**  
1 = Never  
2 = Almost Never  
3 = Sometimes  
4 = Often  
5 = Always

**Sample Questions:**

1. You have a friendly talk with your child  
2. You let your child know when he/she is doing a good job with something  
3. You threaten to punish your child and then do not actually punish him/her  
4. You volunteer to help with special activities that your child is involved with (such as sports, scouts, etc.)  
5. You reward or give something extra to your child for obeying you or behaving well  
6. Your child fails to leave a note to let you know where he/she is going  
7. You play games or do other fun things with your child  

#### üß† Strengths and Difficulties Questionnaire (SDQ) ‚Äì Format & Scoring

The SDQ is a brief behavioral screening tool that includes emotional and behavioral traits. Each question is scored using a 3-point scale:

**Standard Answer Scale:**  
0 = Not True  
1 = Somewhat True  
2 = Certainly True  

Some positive traits are reverse-scored, e.g.:  
- "Generally well behaved, usually does what adults request" ‚Üí 2 = Not True, 0 = Certainly True  
- "Has at least one good friend" ‚Üí 2 = Not True, 0 = Certainly True  
- "Good attention span, sees chores or homework through to the end" ‚Üí 2 = Not True, 0 = Certainly True  

**Impact Supplement Questions:**
- Difficulties present: `0‚Äì3` (None to Severe)
- Duration of difficulties: `0‚Äì3` (Less than a month to Over a year)
- Emotional burden on child or family: `0‚Äì3` (Not at all to A great deal)

**Example SDQ Items:**
- Considerate of other people's feelings
- Often loses temper
- Many worries or often seems worried
- Has at least one good friend
- Often unhappy, depressed or tearful
- Gets along better with adults than other children
- Easily distracted, concentration wanders
- Steals from home, school or elsewhere
- Good attention span, sees chores through to the end

These items are grouped into subscales:
- Emotional Symptoms
- Conduct Problems
- Hyperactivity/Inattention
- Peer Relationship Problems
- Prosocial Behavior

**Total Difficulties Score** = sum of all scales except Prosocial.

The SDQ contributes both individual items and subscale scores to the modeling dataset.

If we don‚Äôt handle missing values in the **test data**, here‚Äôs what can go wrong:

1. Prediction Failure:
Most machine learning models (e.g., XGBoost, RandomForest, LogisticRegression) will raise an error if you try to .predict() with missing values (NaN) in the input.

2. Inconsistent Input Features:
If you imputed (filled) training data but leave NaNs in test data, the model was never trained to expect missing values in those spots ‚Äî this can lead to unreliable or invalid predictions.

3. Skewed Results:
Even if your model handles missing values (like some XGBoost settings do), it may treat NaN as meaningful, which can lead to biased predictions.

- train_df.mean(numeric_only=True) returns a Series where the index is column names and the values are the means from the training data.

- When you apply .fillna() on test_df using that Series, Pandas matches the column names.

- For each column in test_df, if it has a NaN and the same column exists in train_df.mean(), it fills that NaN with the training mean for that column.

### Feature Selection

#### Logistic Regression for Feature Selection

Logistic Regression i s a statistical method used for predicting the probability of a binary outcome (e.g. yes/no, 1/0) based on one or more input features. It's widely used in classification problems.

Why use Logistic Regression for Feature Selection?
LR not only predicts outcomes, but it also provides useful information about how each feature affects the prediction.

How does LR help in feature selection?

- Coefficients: each feautere in the model has a coedficient that indicates how strongly that feature is related to the outcome. Larger absolute values of coefficients indicate more important features.
  - Fit the model
  - Extract coefficients
  - Select top features
- Regularization: Using L1 regularization can further improve feature selection by shrinking less important coefficients to zero.
  - Train the model with L1 Regularization
  - Select non-zero features: after training, features with non-zero coefficients are considered important. These are the features you want to keep for prediction.

**Recursive Feature Elimination (RFE)** is a feature selection technique that involves training a model on a subset of the features, and then iteratively removing the least important features one by one until we are left with the desired number of features.

To implement RFE in python, we can use the RFE class from the sklearn.feature_selection module

Using coefficients:

- Fit a logistic regression model: Train a logistic regression model on your dataset.
- Extract coefficients: the absolute values of the coefficients indicate the importance of the corresponding features.
- Select top features: choose the features with the highest absolute coefficient values.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel

# L1-regularized logistic regression for ADHD_Outcome
log_reg_adhd = LogisticRegression(penalty='l1', solver='liblinear', random_state=42, max_iter=1000)
log_reg_adhd.fit(X_train_resampled, Y_train_resampled['ADHD_Outcome'])

model_adhd = SelectFromModel(log_reg_adhd, prefit=True)
selected_features_adhd = X_train_resampled.columns[model_adhd.get_support()].tolist()

# L1-regularized logistic regression for Sex_F
log_reg_sex = LogisticRegression(penalty='l1', solver='liblinear', random_state=42, max_iter=1000)
log_reg_sex.fit(X_train_resampled, Y_train_resampled['Sex_F'])

model_sex = SelectFromModel(log_reg_sex, prefit=True)
selected_features_sex = X_train_resampled.columns[model_sex.get_support()].tolist()

# Take the intersection of selected features
common_features = list(set(selected_features_adhd) & set(selected_features_sex))
print("Number of common selected features:", len(common_features))

# Filter X datasets based on common features
X_train_selected = X_train_resampled[common_features]
X_test_selected = X_test[common_features]
```

## XGBoost

XGBoost, short for eXtreme Gradient Boosting, is an optimized implementation of gradient boosting, a ml technique that builds an ensemble of decision trees to make predictions. It is widely used for structured (tabular) data and is especially popular in competitive data science (e.g. Kaggle competitions) due to its high performance, speed, and flexibility.

XGBoost builds an ensemble of decision trees in a sequential manner, where each tree corrects the errors of the previous ones. The final prediction is the weighted sum of the predictions from all trees.

Gradient Boosting:

Each tree tries to minimize the error (residual) from the previous step by fitting to the gradient of the loss function. This iterative process improves the model's accuract over time. Ensemble Learning:

By combining multiple weak learners (shallow decision trees), XGBoost creates a strong predictive model.

First we must create an instance of the XGBClassifier from the xgboost library.

Key parameters explained:
- objective = 'binary:logistic': Specifies the task type as binary classification, with logistic regression as the objective function.
- n_estimators=100: sets the number of boosting rounds (i.e. the number of trees to train in the ensemble).
- learning_rate = 0.1: determines the step size shrinkage used to prevent overfitting. Lower values result in slower learning but can improve accuracy.
- max_depth = 5: Limits the depth of each decision tree to prevent overfitting while ensuring the model captures enough complexity.

```python
from xgboost import XGBClassifier
from sklearn.multioutput import MultiOutputClassifier

# Initialize the base classifier
xgb_classifier = XGBClassifier(
    objective='binary:logistic',
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    use_label_encoder=False,
    eval_metric='logloss'  # to suppress deprecation warnings
)
```

Why it's needed: XGBClassifier on its own can only handle single-target classification. MultiOutputClassifier manages the training process by creating a separate XGBClassifier for each target variable in y_train.

```python
# Wrap in MultiOutputClassifier
multioutput_classifier = MultiOutputClassifier(xgb_classifier)
```

### Model Evaluation - Cross Validation Score

The cross-validation score is a metric used to evaluate the performance of a ml model by testing it on multiple subsets of the data It provides an estimate of how well the model is likely to generalize to unseen data.

How Cross-Validation works:
- The dataset is divided into k equally-sized subsets or folds.
- The model is trained on k-1 folds (the training set) and evaluated on the remaining fold (the validation set)
- This process is repeated k times, with each fold being used once as the validation set
- The evaluation metric (we will use accuracy) is computed for each iteration

Why are we using cross-validation score:
- Since we don't have the true lavels for the test data, we cannot calculate the actual accuracy score, Cross-validation provides a reliable way to evaluate our model's performance using the training data without needing the test set answers.
- Cross-validation ensures that a model is evaluated on multiple subsets of the data.

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, accuracy_score
```

Define a custom accuracy metric for multi-output classification, where each target variable (column) needs a separate accuracy calculation.

This function will:
- Converts y_true and y_pred into NumPy arrays to enable indexing like [:, i]
- Iterates over each column (i.e., target variable) in y_true and y_pred.
- Computes the accuracy score for each target variable using accuracy_score.
- Returns the average accuracy across all target variables

```python
def multi_output_accuracy(y_true, y_pred):
  # Ensure y_true and y_pred are numpy arrays
  y_true = np.array(y_true)
  y_pred = np.array(y_pred)
  # Compute accuracy for each target variable and return the mean
  return np.mean([accuracy_score(y_true[:, i], y_pred[:, i]) for i in range(y_true.shape[1])])
```

Then we will wrap the multi_output_accuracy function into a scorer object that can be used with scikit-learn's evaluation tools (e.g., cross_val_score). This will automatically create the y_true and y_pred being used in our function above.

```python
# Crete a scorer using scikit-learn's make_scorer
multi_output_scorer = make_scorer(multi_output_accuracy)
```

Now we will actually perform cross validation with out multi-outcome ml model while using our X_train and Y_train datraframes.

- cv=5: 5 subsets
- scoring = multi_output_scorer: uses the custom scoring function for evaluation

```python
# Note: it takes 12 min to run this cell!
# Perform cross-validation on the training data
cv_scores = cross_val_score(multioutput_classifier, X_train_selected, Y_train_resampled, cv=5, scoring=multi_output_scorer)

# Output the cross-validation results
print("Cross-validation scores for each fold:", cv_scores)
print("Mean CV score:", np.mean(cv_scores))

# Cross-validation scores for each fold: 0.82 0.78 0.69 0.64 0.33
# Mean CV score: 0.6567748188960311
```
